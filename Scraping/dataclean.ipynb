```python
import pandas as pd
```

```python
berk = pd.read_csv('opencourse/berkTest.csv')
cmu = pd.read_csv('opencourse/CMUTest.csv')
corn = pd.read_csv('opencourse/cornellTest.csv')
sail = pd.read_csv('opencourse/SAILTest.csv')
uiuc = pd.read_csv('opencourse/UIUCTest.csv')
yale = pd.read_csv('opencourse/YaleTest.csv')
mit = pd.read_csv('opencourse/mitTest.csv')
stan = pd.read_csv('opencourse/stanfordTest.csv')

allData = [berk, cmu, corn, sail, uiuc, yale, mit, stan]
```

```python
total = 0
for df in allData:
    print(df.columns)
    print()
```

```python

allData[0].drop(columns = ['prerequisites', 'type', 'profs', 'text_profs', 'units'], inplace = True)
allData[0].rename(columns={"websites": "website"}, inplace=True)
allData[0].dropna(subset = ["website"], inplace=True)
allData[0]['college'] = 'UC Berkley'
allData[0].head()
```

```python
allData[1].drop(columns = ['background', 'key_topics', 'last_offered', 'level', 'units', 'meta', 'permission', 'prereq'], inplace = True)
allData[1].dropna(subset = ["website"], inplace=True)
allData[1]['college'] = 'Carnegie Mellon'
allData[1].head()
```

```python
def courseNum(df):
    return df['name'].split(':')[0]
def courseName(df):
    return df['name'].split(':')[len(df['name'])]
def courseName(df):
    return df['name'].split(':')[1]
def courseLink(df):
    l =  eval(df['website'])[0]
    l = list(l.values())
    if len(l) == 0:
        return ''
    return l[0]
allData[2].rename(columns={"sites": "website"}, inplace=True)
allData[2].dropna(subset = ["website"], inplace=True)
allData[2]['number'] = allData[2].apply(courseNum, axis = 1)
allData[2]['name'] = allData[2].apply(courseName, axis = 1)
allData[2]['website'] = allData[2].apply(courseLink, axis = 1)
allData[2]['college'] = 'Cornell'
allData[2].head()
```

```python
allData[3].drop(columns = ['profs'], inplace = True)
allData[3].rename(columns={"link": "website"}, inplace=True)
allData[3].dropna(subset = ["website"], inplace=True)
allData[3]['college'] = 'Stanford'
allData[3].head()

```

```python
allData[4].drop(columns = ['link','prereq','directors','last_updated'], inplace = True)
allData[4].dropna(subset = ["website"], inplace=True)
allData[4]['college'] = 'University of Illinois at Urbana-Champaign'
allData[4].head()
```

```python
allData[5].drop(columns = ['date','professor'], inplace = True)
allData[5].rename(columns={"link": "website"}, inplace=True)
allData[5].dropna(subset = ["website"], inplace=True)
allData[5]['college'] = 'Yale'
allData[5].head()
```

```python
allData[6].drop(columns = ['features','level','taught_in'], inplace = True)
allData[6].rename(columns={"link": "website"}, inplace=True)
allData[6].dropna(subset = ["website"], inplace=True)
allData[6]['college'] = 'Massachusetts Institute of Technology'
allData[6].head()
```

```python
allData[7].drop(columns = ['prof'], inplace = True)
allData[7].rename(columns={"link": "website"}, inplace=True)
allData[7].dropna(subset = ["website"], inplace=True)
allData[7]['college'] = 'Stanford'
allData[7].head()
```

```python

combined = pd.merge(allData[0], allData[1], how="outer")
combined = pd.merge(combined, allData[2], how="outer")
combined = pd.merge(combined, allData[3], how="outer")
combined = pd.merge(combined, allData[4], how="outer")
combined = pd.merge(combined, allData[5], how="outer")
combined = pd.merge(combined, allData[6], how="outer")
combined = pd.merge(combined, allData[7], how="outer")
combined
combined.to_csv('cleanData.csv')
```
